# Introducción a la librería *Pandas*

`Pandas` es una muy popular librería de código abierto dentro de los desarrolladores de *Python*, y sobre todo, junto con `Numpy` y `sckit`, dentro del ámbito de **Data Science** y **Machine Learning**. Ofrece unas estructuras muy poderosas y flexibles que facilitan la manipulación y tratamiento de datos.

¿Qué nos permite hacer *Pandas*?

  * Trabajar con bases de datos, por ejemplo las calificaciones del grupo de Métodos Cuantitativos en Finazas, para lo cual usamos datos de tipo `DataFrame`.
  * Trabajar con datos indexados respecto a otra variable (solo una variable), como el precio de cierre de una acción en el tiempo, para lo cual usamos los datos de tipo `Series`.
  * Con *Pandas* podemos acceder y manipular datos dentro de `DataFrame` y `Series`.
  * Podemos importar datos csv, .xlsx, descargar información de la nube, de internet o directamente crearlos desde Python.

Para ello, debemos importar la librería `pandas` almacenándola bajo el alias `pd`, bajo la siguiente línea de código.

````python
import pandas as pd
````

 Las estructuras de datos principales en *pandas* están implementadas en dos clases:

  * **`DataFrame`**, que podemos interpretar como *una tabla de datos relacional*, con filas y columnas con nombre.
  * **`Series`**, que es una columna simple. Una clase `DataFrame` incluye una o más `Series` y un nombre para cada `Series`.

  <img src='https://drive.google.com/uc?export=view&id=14lY_hN5sbf6knIVkSMzEEVcGz79fJWWL' width=500px>

 Una manera de crear una `Series` es construir un objeto de `Series`. Por ejemplo:

````python
pd.Series(['Mau', 'Mayte', 'Jair', 'Francisco'])
````

Los objetos tipo `DataFrame` pueden crearse al enviar una variable tipo `dict` (diccionario) que asigne nombres de columnas de `string` a sus `Series` correspondientes. Si las `Series` no coinciden con la longitud, los valores que falten se completan con valores NA/NaN especiales. Ejemplo:

````python
alumnos = pd.Series( ['Mau', 'Mayte', 'Jair', 'Francisco'] )
calificaciones = pd.Series([10.15, 10.9, 10.0, 10.21])

pd.DataFrame( { 'Alumno': alumnos, 'Calificacion': calificaciones } )
````

Por lo general, nosotros trabajamos con datos precargados, ya sea desde un archivo completo en nuestra computadora o bien descargándolo de internet o la nuebe. 

El siguiente ejemplo carga un archivo con datos de viviendas de California.

````python
california_df = pd.read_csv("https://download.mlcc.google.com/mledu-datasets/california_housing_train.csv", sep=",")
````

Una forma de visualizar los datos de nuestro `DataFrame` podría ser llamando directamente a la variable creada. Otra forma más simplificada de hacerlo es con el método `.head()`, que muestra los primeros registros de un `DataFrame`. 

Si omitimos el parámetro que recibe, imprimirá los primeras cinco observaciones (filas) del `Dataframe`. 

Así, el método `.head(n)` imprimirá las primeras cinco observaciones del `DataFrame`.



````python
california_df.head()
````

````python
california_df.head(11)
````

El método `.tail()`, análogamente al método anterior, nos imprime las últimas observaciones del `DataFrame`.

````python
california_df.tail(11)
````

Uno de los métodos más utilizados en la manipulación de `DataFrames` es el métodos `.describe()`.

Este método nos brinda estadísticas importantes sobre las variables/columnas que tenemos en un `DataFrame`.

````python
california_df.describe()
````

Si quisiéramos modificar los nombres de las columnas/variables de nuestro `DataFrame`, podemos hacerlo mediante el método `.columns()`.

````python
list(california_df.columns)
````

````python
california_df.columns = ['Longitud', 
                         'Latitud', 
                         'Mediana de la edad', 
                         'Cuartos totales', 
                         'Habitaciones totales', 
                         'Poblacion', 'Hogares', 
                         'Mediana del ingreso', 
                         'Mediana del valor de la casa']
california_df.head()
````

Si quisiera cambiar únicamente el nombre de una o algunas columnas usamos el método `.rename()`. Para ello, pasamos un diccionario con keys los nombres originales y values los nuevos nombres.

````python
california_df.rename(columns = {'Mediana de la edad' : 'Mediana edad', 
                                'Mediana del ingreso': 'Mediana ingreso', 
                                'Mediana del valor de la casa':'Mediana valor casa'}, 
                     inplace = True)
california_df.head(3)
````

 Otra función util de *pandas* es la generación de gráficos. Por ejemplo, `.hist()` permite estudiar rápidamente la distribución de los valores en una columna:

Existen varias propiedades o métodos para ver las características de un `DataFrame`. A continuación resumimos algunas de las más usadas en la práctica

*`df.info()` : Devuelve información (número de filas, número de columnas, índices, tipo de las columnas y memoria usado) sobre el `DataFrame` df.

*`df.shape` : Devuelve una tupla con el número de filas y columnas del `DataFrame` df, es decir una tupla (filas, columnas).

*`df.size` : Devuelve el número de elementos del `DataFrame`.

*`df.columns` : Devuelve una lista con los nombres de las columnas del `DataFrame` df.

*`df.index` : Devuelve una lista con los nombres de las filas del `DataFrame` df.

*`df.dtypes` : Devuelve una serie con los tipos de datos de las columnas del `DataFrame` df.

## Manipulación y filtrado de datos con *Pandas*

Pandas nos permite obtener columnas o filas de nuestros datos de forma muy fácil e intuitiva. Además, podemos hacer una exploración basándonos en condiciones tal como veremos a continuación.

De forma análoga a como accedíamos a datos de una lista, podemos hacer uso del operador `[]` para acceder únicamente a una columna. Por ejemplo:

````python
california_df['Poblacion']
````

````python
california_df.shape # (filas, columnas) o bien (observaciones, variables) 
                    # PARA RECORDARLO (Y,X)
````

El acceso a los datos de un `DataFrame` se puede hacer a través de posiciones o través de los nombres de las filas y columnas.
1.   `.loc[]`, por **nombre** de fila.
2.   `.iloc[]`, por el **índice** de nuestra fila.

### Método `.loc[]` (*location*)

`df.loc[fila, columna]` : Devuelve el elemento que se encuentra en la fila con nombre `fila` y la columna con nombre `columna` del DataFrame df.

````python
california_df.head()
````

````python
california_df.loc[3, 'Poblacion']
````

````python
california_df.loc[3, ['Longitud', 'Latitud', 'Poblacion'] ]
````

````python
california_df.loc[3 , list(california_df.columns) ]
````

<br><font color = #3F6E9A size=5> NOTA PARA .loc: </font> El operador `inicio:fin` *toma* ambos extremos, en el sentido que se imprimen tanto el valor inicial como el valor final.

Si se omite `inicio`, es decir algo tipo `:fin`, por default se hará `0:fin`

Si se omite `fin`, es decir algo tipo `inicio:`, por default se hará `inicio:len(df)`

````python
california_df.loc[:7, ['Mediana edad', 'Cuartos totales', 'Poblacion'] ]
````

````python
california_df.loc[10:20, ['Mediana edad', 'Cuartos totales', 'Poblacion'] ]
````

### Método `.iloc[]` (*indexed location*)

* `df.iloc[i, j]` : Devuelve el elemento que se encuentra en la fila `i` y la columna `j` del `DataFrame` llamado `df`. Pueden indicarse secuencias de índices para obtener partes del `DataFrame`.

````python
california_df.head(4)
````

Con el método `.loc[]`

````python
california_df.loc[3, 'Poblacion']
````

Con el método `.iloc[]`

````python
california_df.iloc[3, 5]
````

<br><font color = #3F6E9A size=5> NOTA PARA .iloc: </font> El operador `inicio:fin` *toma* solo el primer extremo, de forma similar a como ocurría con `range(inicio,fin)`

Si se omite `inicio`, es decir algo tipo `:fin`, por default se hará `0:fin`

Si se omite `fin`, es decir algo tipo `inicio:`, por default se hará `inicio:len(df)`

````python
for n in range(10,15):
  print(n)
````

Con el método `.loc[]`

````python
california_df.loc[:7, ['Mediana edad', 'Cuartos totales', 'Habitaciones totales', 'Poblacion'] ]
````

Con el método `.iloc[]`

````python
california_df.iloc[:7, 2:6]
````

Para filtrar datos dentro de un `DataFrame` haremos uso de las *sentencias booleanas*. Cuando trabajemos con un `DataFrame` vamos a trabajar con datos que cumplan la sentencia establecida (`True`)



````python
california_df[ california_df['Poblacion'] > 1500].head(3)
````

````python
california_df[ california_df['Poblacion'] > california_df['Poblacion'].mean()].head(3)
````

````python
california_df[ (california_df['Poblacion'] > 1500) & (california_df['Mediana edad'] == 15.0) ].head(5)
````

Podemos también *aplicar funciones sobre las columnas de un `DataFrame`*. Para ello hacemos uso del método `apply()`.


```
df[colum].apply(function)
```
De esta forma aplicamos la función `function` a la columna llamada `colum`, del `DataFrame` llamado `df`


````python
import math

california_df['Log(Poblacion)'] = california_df['Poblacion'].apply(math.log)
california_df.head(3)
````

*`df.append(serie, ignore_index=True)` : Devuelve el `DataFrame` que resulta de añadir una fila al `DataFrame` llamado `df` con los valores de la Serie `serie`. Los nombres del índice de la serie deben corresponderse con los nombres de las columnas de `df`. 

Si no se pasa el parámetro `ignore_index` entonces debe pasarse el parámetro `name` a la serie, donde su argumento será el nombre de la nueva fila.

````python
dato_cero = california_df.iloc[0,]
list(dato_cero)
````

````python
list(california_df.columns)
````

````python
california_df = california_df.append(pd.Series ( list(dato_cero), index = list(california_df.columns) ) , ignore_index = True)
california_df.tail()
````

Para ordenar los valores de un `DataFrame` usamos el método `.sort_values()`


```
df.sort_values(colum, ascending = booleano) 
```
Devuelve el `DataFrame` que resulta de ordenar las filas del `DataFrame` llamado `df` según los valores del la columna con nombre `colum`. Si argumento del parámetro `ascending` es `True` el orden es creciente y si es `False` decreciente.


````python
california_df.sort_values('Poblacion', ascending = True).head()
````
